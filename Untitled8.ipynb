```python
import random  # Rastgele sayı üretmek için kullanılan Python modülü

# Sürücünün hangi takıma ait olduğunu gösteren sözlük
driver_team = {
    "Max Verstappen":         "Red Bull Racing",
    "Yuki Tsunoda":           "Red Bull Racing",
    "Lando Norris":           "McLaren",
    "Oscar Piastri":          "McLaren",
    "George Russell":         "Mercedes",
    "Kimi Antonelli":         "Mercedes",
    "Charles Leclerc":        "Ferrari",
    "Lewis Hamilton":         "Ferrari",
    "Fernando Alonso":        "Aston Martin",
    "Lance Stroll":           "Aston Martin",
    "Pierre Gasly":           "Alpine",
    "Franco Colapinto":       "Alpine",
    "Alex Albon":             "Williams",
    "Carlos Sainz":           "Williams",
    "Liam Lawson":            "Racing Bulls",
    "Isack Hadjar":           "Racing Bulls",
    "Nico Hülkenberg":        "Sauber",
    "Gabriel Borteletto":     "Sauber",
    "Esteban Ocon":           "Haas",
    "Oliver Baerman":         "Haas"
}

# Takımların güvenilirlik oranı (DNF olasılığı baz katsayısı)
team_reliability = {
    "McLaren": 0.025,
    "Red Bull Racing": 0.030,
    "Mercedes": 0.035,
    "Ferrari": 0.035,
    "Aston Martin": 0.040,
    "Alpine": 0.045,
    "Williams": 0.050,
    "Racing Bulls": 0.055,
    "Sauber": 0.050,
    "Haas": 0.060
}



# Yarışa katılan tüm pilotların isimlerini içeren liste
drivers = [
    "Max Verstappen", "Lewis Hamilton", "Charles Leclerc", "Lando Norris",
    "Fernando Alonso", "Esteban Ocon", "Pierre Gasly", "Carlos Sainz", "George Russell",
    "Alex Albon", "Lance Stroll", "Yuki Tsunoda", "Oscar Piastri", "Nico Hülkenberg",
    "Kimi Antonelli", "Gabriel Borteletto", "Liam Lawson", "Isack Hadjar",
    "Oliver Baerman", "Franco Colapinto"
]

# Her pilotun yetenek puanı (senin verdiğin değerler korunmuştur)
skill_points = {
    "Max Verstappen": 10,
    "Lewis Hamilton": 8.5,
    "Charles Leclerc": 8.2,
    "Lando Norris": 8.5,
    "Fernando Alonso": 8,
    "Esteban Ocon": 7.1,
    "Pierre Gasly": 6.5,
    "Carlos Sainz": 7,
    "George Russell": 9.2,
    "Alex Albon": 7.6,
    
    "Lance Stroll": 5.2,
    "Yuki Tsunoda": 6.8,
    "Oscar Piastri": 8.7,
    "Nico Hülkenberg": 7,
    "Kimi Antonelli": 7.7,
    "Gabriel Borteletto": 6.8,
    "Liam Lawson": 6.1,
    "Isack Hadjar": 6.95,
    "Oliver Baerman": 6.7,
    "Franco Colapinto": 6.9
}

# Araç güçleri (senin sağladığın sabit değerler)
car_powers = {
    "Max Verstappen": 0.8,
    "George Russell": 0.82,
    "Lewis Hamilton": 0.72,
    "Lando Norris": 0.94,
    "Kimi Antonelli": 0.8,
    "Charles Leclerc": 0.72,
    "Oscar Piastri": 0.94,
    "Isack Hadjar": 0.69,
    "Gabriel Borteletto": 0.68,
    "Fernando Alonso": 0.7,
    "Alex Albon": 0.7,
    "Carlos Sainz": 0.68,
    "Franco Colapinto": 0.68,
    "Oliver Baerman": 0.65,
    "Pierre Gasly": 0.66,
    "Liam Lawson": 0.65,
    "Yuki Tsunoda": 0.70,
    "Nico Hülkenberg": 0.66,
    "Esteban Ocon": 0.66,
    "Lance Stroll": 0.68
}

# Pist süreleri ve tur sayıları (tam liste)
track_base_times = {
    "Bahrain": 94.0, "Saudi Arabia": 91.8, "Australia": 85.3, "Japan": 89.4,
    "China": 91.2, "Miami": 92.6, "Imola": 90.0, "Monaco": 79.6, "Canada": 83.9,
    "Spain": 89.5, "Austria": 66.3, "Britain": 87.3, "Hungary": 80.4,
    "Belgium": 110.5, "Netherlands": 72.5, "Italy": 76.2, "Azerbaijan": 98.0,
    "Singapore": 105.0, "USA": 93.0, "Mexico": 78.3, "Brazil": 71.0,
    "Las Vegas": 100.0, "Qatar": 89.8, "Abu Dhabi": 96.2
}
race_laps = {
    "Australia": 58, "China": 56, "Japan": 53, "Bahrain": 57, "Saudi Arabia": 50,
    "Miami": 57, "Imola": 63, "Monaco": 78, "Spain": 66, "Canada": 70, "Austria": 71,
    "Britain": 52, "Belgium": 44, "Hungary": 70, "Netherlands": 72, "Italy": 53,
    "Azerbaijan": 51, "Singapore": 61, "USA": 56, "Mexico": 71, "Brazil": 71,
    "Las Vegas": 50, "Qatar": 57, "Abu Dhabi": 58
}

# Hava koşulları ve etki
weather_conditions = {
    "Sunny": 0,
    "Cloudy": 0.5,
    "Rainy": 0  # random.uniform içinde belirlenecek
}

# Sezon puanları başlat
player_points = {driver: 0 for driver in drivers}

def wait_for_enter ():
    input("\n>> Devam etmek icin ENTER'a bas <<")

def print_standings():
    print("\n📊 Sürücü Şampiyonası Puan Durumu:")
    sorted_points = sorted(player_points.items(), key=lambda x: x[1], reverse=True)
    for i, (driver, pts) in enumerate(sorted_points, 1):
        print(f"{i:2}. {driver:<20} {pts} puan")
        
def print_dnfs(lap_times):
    dnf_list = [d for d, t in lap_times.items() if t is None]
    if dnf_list:
        print("\n❌ DNF OLANLAR:")
        for d in dnf_list:
            print(f" - {d:<20} (skill: {skill_points[d]})")




def simulate_race(track):
    laps = race_laps[track]
    base = track_base_times[track]
    weather_today = random.choice(list(weather_conditions.keys())) 
    # Hava cezası
    weather_penalty = random.uniform(8,10) if weather_today=="Rainy" else weather_conditions[weather_today]

    lap_times = {}
    dnfs = []
    
    for driver in drivers:

        team = driver_team[driver]
        base_dnf = team_reliability.get(team, 0.05)
        aggression = (10 - skill_points[driver]) * 0.005
        dnf_chance = min(0.99, base_dnf + aggression)

        # DNF kontrolü
        if random.random() < dnf_chance:
            dnfs.append(driver)
            lap_times[driver] = None
            
            continue
        # Süre hesaplama
        skill = skill_points[driver]
        car_effect = (0.5 - car_powers[driver]) * 20
        rand = random.uniform(-0.3, 0.3)
        adjusted = base + rand + car_effect - skill * 0.8
        if adjusted < 70:
            adjusted = 70
        pits = laps // 25
        pit_penalty = sum(random.uniform(18,23) for _ in range(pits))
        total = adjusted * laps + pit_penalty + weather_penalty
        lap_times[driver] = total

    # Tamamlayanları sırala
    finished = [(d,t) for d,t in lap_times.items() if t is not None]
    sorted_results = sorted(finished, key=lambda x: x[1])
    # Puan dağıtım
    points = [25,18,15,12,10,8,6,4,2,1]
    puan_sayisi = min(10, len(sorted_results))

    for i in range(puan_sayisi):
        driver, _ = sorted_results [i]
        player_points[driver] += points [i]
    return lap_times, sorted_results, weather_today

# Örnek yarış çalıştırma
season_tracks = list(track_base_times.keys())


for round_no, track in enumerate(season_tracks, 1):
    input(f"\n>> ENTER'a basarak {round_no}. YARIŞ ({track.upper()})'ı başlat <<")
    print(f"\n{'='*40}\n{round_no}. YARIS _=- {track.upper()} GRAND PRIX\n{'='*40}")
    lap_times , results , weather = simulate_race(track)

    print(f"Hava Durumu: {weather}")
    print("\nYarış Sonuçları:")
    for idx , (driver, sec) in enumerate (results, start=1):
        mn = int(sec) // 60
        sc = sec % 60
        print(f"{idx:2}. {driver:<20} {mn}:{sc:06.3f} (skill: {skill_points[driver]})")
        


# DNF olanlar
    print_dnfs(lap_times)
    print_standings()
print("\n 🏆 SEZON BİTTİ! ")


print("\nSürücü Şampiyonası Puan Durumu")
sorted_points = sorted(player_points.items(), key=lambda x: x[1], reverse=True)
for i, (driver , pts) in enumerate(sorted_points, 1 ):
    print(f"{i:2}. {driver:20} {pts} puan")

input("\n>> Devam etmek için ENTER'a bas <<")
```
